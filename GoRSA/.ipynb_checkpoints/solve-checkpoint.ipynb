```python
# 毎回cやdを打ち込みたくない人向け
import math
from socket import *
from Crypto.Util.number import long_to_bytes

s = socket(AF_INET, SOCK_STREAM)
s.connect(("133.242.17.175", 1337))
# c
c = s.recv(700)
# 先頭の'Encrypted flag is: 'と末尾の'\n> 'を削除
c = int(c[19:-3])

#y1
s.send(b'2\n')
y1 = s.recv(700)
y1 = int(y1[:-3])

#y2
s.send(b'4\n')
y2 = s.recv(700)
y2 = int(y2[:-3])

#y3
s.send(b'16\n')
y3 = b''
# 改行まで読み取る
while True:
    y3chr = s.recv(1)
    if y3chr == b'\n':
        y3 = int(y3)
        break
    y3 += y3chr

# d
d = s.recv(700)
# 先頭の'The D was 'と末尾の'\n'を削除
d = int(d[10:-1])

# 以下同じ
kN = math.gcd(pow(y1,2) - y2, pow(y2,2) - y3)
k = 1
y_max = max(y1,y2,y3)
while kN // k >= y_max:
    N = kN // k
    print(long_to_bytes(pow(c, d, N)))
    k += 1
```
